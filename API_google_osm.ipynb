# Para sacar calles limpias

```python
import requests
import csv

provinces = ['Murcia', 'Almeria', 'Alicante', 'Valencia', 'Pontevedra', 
             'Toledo', 'Madrid', 'Valladolid', 'Las Palmas', 'A Coruna', 
             'Barcelona', 'Asturias']

overpass_url = "http://overpass-api.de/api/interpreter"

with open('streets.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["name", "province"]) # writing headers

    for province in provinces:
        overpass_query = f"""
        [out:csv("::id", "::type", "name";false)];
        area[name="{province}"]->.a;
        way(area.a)[highway][name];
        out;
        """
        response = requests.get(overpass_url, params={'data': overpass_query})
        data = response.text.split('\n')[1:] # split response by newlines and remove header row

        for row in data:
            if row: # check row is not empty
                writer.writerow(row.split('\t')[2:] + [province]) # split each row by tabs, exclude id and type, and append province
```
